# For Saree

```python
# Lets first connect to the web driver

driver=webdriver.Chrome(r"C:\Users\Visual singh\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
url = "https://www.amazon.in/"
driver.get(url)
```

```python
input_ = driver.find_element(By.XPATH,"//input[@id='twotabsearchtextbox']")
input_.send_keys("sarees")
```

```python
btn = driver.find_element(By.XPATH,"//input[@id='nav-search-submit-button']").click()
```

```python
images_saree = []
start=0
end=3

for page in range(start,end):#for loop for scrapping 3 page
    
    elements = driver.find_elements(By.XPATH,"//img[@class='s-image']")
    for i in elements:
        image = i.get_attribute('src')
        images_saree.append(image)

for img in images_saree:
    file_name = img.split('/')[-1]
    print(f"this is the saree name:{file_name}")
    r = requests.get(img, stream=True)
    if r.status_code == 200:
        with open(file_name, 'wb') as f:
            for chunk in r:
                f.write(chunk)
                
    else:
        print('')
        
    nxt_button=driver.find_elements(By.XPATH,"//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")#scraping the list of buttons from the page
```

```python
Sarees=pd.DataFrame({'Saree':images_saree})
Sarees
```

```python
print(len(Sarees))
```

```python
driver.close()
```
